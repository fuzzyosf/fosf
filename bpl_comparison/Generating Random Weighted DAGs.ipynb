## Generating a random weighted DAG with the same defaults used in the fosf and Bousi~Prolog comparison

```python
import numpy as np
import random

from bpl import generate_random_dag
from fosf.utils.draw import notebook_display as display
from fosf.syntax import FuzzySortTaxonomy
```

```python
%time
G = generate_random_dag(10, initial_branches=3, base_nodes_per_branch=2, split_probability=0.05, intra_branch_p=0.1, seed=1)
random_degree = lambda: round(np.random.uniform(0.1, 1), 3)
fuzzy_edges = [(f"s{v}", f"s{u}", random_degree()) for u, v in G.edges()]
```

```python
%%time
ftax = FuzzySortTaxonomy(fuzzy_edges)
```

```python
ftax.graph.number_of_nodes()
```
